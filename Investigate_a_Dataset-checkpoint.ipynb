### Renaming columns

```python

edited_df=df.rename({'Hipertension': 'Hypertension','No-show':'Missed_appointment','Neighbourhood':'Hospital_location'},axis=1)
edited_df.head()
```

### Computing the number of people who missed their appointment and those who honored their appointment

```python
declined=sum(edited_df['Missed_appointment']=='Yes')
print(declined, "people missed their appointment")


honored=sum(edited_df['Missed_appointment']=='No')
print(honored, "people did not miss their appointment")
```

### Creating two sub dataframes df_a for those who showed up for appointment and df_b for those who did not show up

```python
df_a=edited_df.query('Missed_appointment == "No"') 
df_a
```

```python
df_b=edited_df.query('Missed_appointment == "Yes"')
df_b
```

### Converting the values of Missed_appointment column 

### 'Yes' becomes 0 for 0 appointment attended, and 'No' becomes 1, for an appointment attended.

```python
showed_up=np.repeat(1,df_a.shape[0])
```

```python
Did_not_show=np.repeat(0,df_b.shape[0])
```

### Creating a new column "Attended"

```python
df_a['Attended']=showed_up
df_a.head()
```

```python
df_b['Attended']=Did_not_show
df_b.head()
```

## Appending the two sub dataframes

```python
second_edit_df=df_a.append(df_b)
second_edit_df.head()
```

## Exploratory Data Analysis

### Histograms of various variables

```python
second_edit_df.hist(figsize=(15,15));
```

### Visualization for age counts

```python
second_edit_df['Age'].value_counts().plot(kind='bar',figsize=(25,25),fontsize=(14));
plt.xlabel('Age',fontsize=(25))
plt.ylabel('Counts',fontsize=(25));
```

### Research Question 1 (How does age influence the probability of showing up)


### Mean attendance by age

```python
second_edit_df.groupby('Age')['Attended'].mean()
```

```python
second_edit_df.groupby('Age')['Attended'].mean().plot(kind='bar', figsize=(25,25),fontsize=(14));
plt.title('Attendance with consideration to age',fontsize=(25))
plt.xlabel('Age',fontsize=(25))
plt.ylabel('Showed up for the appointment',fontsize=(25));
```

### Research Question 2  (Does gender contribute to the probability of showing up)

### Mean attendance by gender

```python
second_edit_df.groupby('Gender')['Attended'].mean()
```

```python
second_edit_df.groupby('Gender')['Attended'].mean().plot(kind='bar',alpha=.7);
plt.title('Attendance based on gender')
plt.xlabel('Gender')
plt.ylabel('Showed up for the appointment');
```

### Research Question 3(Is the location of a hospital a factor affecting the probability of showing up)

### Mean attendance by hospital location

```python
second_edit_df.groupby('Hospital_location')['Attended'].mean()
```

```python
second_edit_df.groupby('Hospital_location')['Attended'].mean().plot(kind='bar',figsize=(20,20),fontsize=(12));
plt.title('Attendance by location of hospital',fontsize=(25))
plt.xlabel('Hospital_location',fontsize=(25))
plt.ylabel('Showed up for the appointment',fontsize=(25));
```

### Research Question 4 (Does having the scholarship or not, affect the probability of showing up)

### Mean attendance by scholarship

```python
second_edit_df.groupby(['Scholarship','Attended']).mean()
```

```python
second_edit_df.groupby('Scholarship')['Attended'].mean().plot(kind='bar',alpha=.7);
plt.title('Attendance based on scholarship')
plt.xlabel('Status of scholarship')
plt.ylabel('Showed up for the appointment');
```

### Research Question 5 (Does receiving an SMS influence the probability of showing up)

### Mean attendance if SMS received

```python
second_edit_df.groupby('SMS_received')['Attended'].mean()
```

```python
second_edit_df.groupby('SMS_received')['Attended'].mean().plot(kind='bar',alpha=.7);
plt.title('Attendance based on receiving SMS')
plt.xlabel('Received SMS?')
plt.ylabel('Showed up for the appointment');
```

### Research Question 6 (How does level of being handicap affect the probability of showing up)

## Mean attendance if patient is handicap

```python
second_edit_df.groupby('Handcap')['Attended'].mean().plot(kind='bar',alpha=.7);
plt.title('Attendance by degree of handcap')
plt.xlabel('Handcap')
plt.ylabel('Showed up for appointment');
```
